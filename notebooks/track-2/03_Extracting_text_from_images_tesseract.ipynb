### Imagen de ejemplo

```python
image_path = 'example_script_03.jpg'

if not os.path.exists(image_path):
    sample_image = Image.new('RGB', (1200, 700), 'white')
    draw = ImageDraw.Draw(sample_image)

    try:
        title_font = ImageFont.truetype('DejaVuSans-Bold.ttf', 48)
        body_font = ImageFont.truetype('DejaVuSans.ttf', 36)
    except OSError:
        title_font = ImageFont.load_default()
        body_font = ImageFont.load_default()

    draw.rectangle((40, 40, 1160, 420), outline='black', width=4)
    draw.text((70, 75), 'Track 2 OCR Demo', fill='black', font=title_font)
    draw.multiline_text(
        (70, 175),
        'Imagen a texto\nValidar y limpiar resultado\nLectura de ejemplo',
        fill='black',
        font=body_font,
        spacing=16,
    )
    sample_image.save(image_path)
    print(f"No se encontró {image_path}; se generó una imagen demo para empezar.")

image0 = Image.open(image_path)
image0
```

```python
image_path = 'example_script_03.jpg'

if not os.path.exists(image_path):
    sample_image = Image.new('RGB', (1200, 700), 'white')
    draw = ImageDraw.Draw(sample_image)

    try:
        title_font = ImageFont.truetype('DejaVuSans-Bold.ttf', 48)
        body_font = ImageFont.truetype('DejaVuSans.ttf', 36)
    except OSError:
        title_font = ImageFont.load_default()
        body_font = ImageFont.load_default()

    draw.rectangle((40, 40, 1160, 420), outline='black', width=4)
    draw.text((70, 75), 'Track 2 OCR Demo', fill='black', font=title_font)
    draw.multiline_text(
        (70, 175),
        'Imagen a texto\nValidar y limpiar resultado\nLectura de ejemplo',
        fill='black',
        font=body_font,
        spacing=16,
    )
    sample_image.save(image_path)
    print(f"No se encontró {image_path}; se generó una imagen demo para empezar.")

extracted_information = image_to_string(Image.open(image_path))
extracted_information
```

### Qué observar del resultado

La salida no siempre será perfecta. OCR traduce patrones visuales en texto, pero no “comprende” la imagen.

Si quieres mejorar el resultado, suele ayudar preprocesar la imagen, recortar ruido y revisar manualmente el texto final.
