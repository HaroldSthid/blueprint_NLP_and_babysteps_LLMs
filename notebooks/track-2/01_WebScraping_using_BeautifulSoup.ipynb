```python
# Importamos las herramientas necesarias
from bs4 import BeautifulSoup
from urllib.request import urlopen
```

```python
from urllib.request import Request  # Importamos la clase Request

myurl = 'https://stackoverflow.com/questions/415511/how-to-get-the-current-time-in-python'  # URL de ejemplo
# Creamos un Request con un User-Agent transparente para identificar la solicitud de forma responsable
req_obj = Request(myurl, headers={'User-Agent': 'NLPFundamentalsTrack2LearningBot/1.0'})
req = urlopen(req_obj)  # Enviamos el Request con urlopen
html = req.read()  # Leemos el HTML
soupified = BeautifulSoup(html, 'html.parser')  # Parseamos el HTML con BeautifulSoup
```

Como el HTML de una página puede ser largo, mostramos solo un fragmento para no saturar la salida.


```python
print(soupified.prettify()[:2000])  # Mostramos solo una parte para orientar la lectura
```

Mini reto de depuración: el HTML real cambia con el tiempo. Si un selector falla, revisa la estructura actual, consulta la documentación o apóyate en IA para inspeccionar el DOM y ajustar la extracción de forma segura.


```python
def extraer_texto_post(contenedor, etiqueta):
    if contenedor is None:
        print(f"{etiqueta}: no se encontró el contenedor en esta página.")
        return

    cuerpo = contenedor.find("div", {"class": "s-prose js-post-body"})
    if cuerpo is None:
        print(f"{etiqueta}: se encontró el contenedor, pero falta el bloque de contenido.")
        return

    print(f"{etiqueta}:\n", cuerpo.get_text().strip())

question = soupified.find("div", {"class": "question"})  # Buscamos el contenedor de la pregunta
extraer_texto_post(question, "Pregunta")

answer = soupified.find("div", {"class": "answer"})  # Buscamos una respuesta
extraer_texto_post(answer, "Respuesta destacada")
```

## ¡Claro!

En palabras sencillas, este script hizo cuatro cosas: preparó las herramientas, visitó una página de Stack Overflow, leyó su HTML con BeautifulSoup y extrajo la pregunta junto con la mejor respuesta.

La idea importante es que el encabezado User-Agent ayuda a identificar la solicitud de forma responsable con un identificador claro; sigue siendo necesario respetar las reglas del sitio, sus límites y sus tiempos de espera.
